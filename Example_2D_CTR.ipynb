```python
%reset -f
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
from CTR import *
from CTR.PIticks import PIticks
```

```python
Sr = Atom(32)
Ti = Atom(22)
Ru = Atom(44)
O  = Atom(8)
In = Atom(49)
Sb = Atom(51)

#
sto = Perovskite(
	abc = 3.905 * vec(1, 1, 1),
	ABO = (Sr, Ti, O)
)
sro = Perovskite(
    abc = Molecule.pseudocubic(5.567, 5.5304, 7.8446),
    ABO = (Sr, Ru, O)
)
# insb = Molecule(
#     abc = vec(4.63, 8.01, 7.62),
#     structure = [
        
# 	]
# )
#
STO_Bulk = Film(sto, (inf,inf,inf))
SRO = Film(sro/sto(0,0,1), (inf,inf,10))
```

```python
XRD = Xray()
XRD.Q(1,1,0)
plt.figure(figsize=(12,3))
plt.semilogy(XRD.TTHETA, XRD.I(SRO/STO_Bulk(0,0,1)), linewidth=3, color='#006080')
plt.xlabel('2 (Deg.)', fontdict={'size': 16})
plt.ylabel('Intensity', fontdict={'size': 16})
plt.title("Strained SRO/STO(001),  (110)Scan", fontdict={'size': 16})
```

```python
XRD2D = Xray2d()
XRD2D.Q(0,0,1)
X, Y, I = XRD2D.X, XRD2D.Y, XRD2D.I(STO_Bulk)

# plt.pcolor(X, Y, I, norm=LogNorm(), cmap='gray')
plt.contourf(X, Y, I, norm=LogNorm(vmin=I.min(), vmax=I.max()), cmap='gray')
plt.xlabel('Qx [1/]')
plt.ylabel('Qy [1/]')
plt.title('STO(inf, inf, 10) BULK Intensity')
plt.xticks(*PIticks(0, 2.5, 0.5))
plt.yticks(*PIticks(0, 2.5, 0.5))
plt.colorbar()
```

```python
X, Y, I = XRD2D.X, XRD2D.Y, XRD2D.I(SRO/STO_Bulk(0,0,1))

plt.pcolor(X, Y, I, norm=LogNorm(), cmap='gray')
# plt.contourf(X, Y, I, norm=LogNorm(vmin=I.min(), vmax=I.max()), cmap='gray')
plt.xlabel('Qx [1/]')
plt.ylabel('Qy [1/]')
plt.title('Strained SRO/STO(0,0,1) Intensity')
plt.xticks(*PIticks(0, 2.5, 0.5))
plt.yticks(*PIticks(0, 2.5, 0.5))
plt.colorbar()
```
