```python
import random
```

```python
class Game:
    # 가로 x, 세로 y, 지뢰 mine 개의 게임판을 만든다
    def __init__(self, x, y, mine):
        # 남은 칸 수
        self.left = x * y - mine
        # 남은 지뢰의 개수
        self.mine_left = mine
        # 게임 종료 여부
        self.is_game_over = False
        # 승리 여부
        self.is_win = False
        # 겉으로 보이는 스크린 -2는 코드의 편의를 위해 만든 벽면 -1은 깃발 0은 닫힌 칸 1은 열린 칸 2는 지뢰 (터졌을 때 나옴)
        self.screen = [[-2 for _ in range(x + 2)]] + [[-2] + [0 for _ in range(x)] + [-2] for _ in range(y)] + [
            [-2 for _ in range(x + 2)]]
        # 실제 게임 판 -2는 코드의 편의를 위해 만든 벽면 -1은 지뢰 나머지는 숫자
        self.board = [[-2 for _ in range(x + 2)]] + [[-2] + [0 for _ in range(x)] + [-2] for _ in range(y)] + [
            [-2 for _ in range(x + 2)]]

        # 지뢰를 mine 개 집어 넣어 준다
        for i in random.sample(range(x * y), mine):
            self.board[(i // x + 1)][(i % x) + 1] = -1

        # 주변 8칸의 상대적인 좌표를 저장해주는 문자열
        around = "0001021012202122"
        # 모든 칸을 돌면서 각 칸 주변에 있는 지뢰의 개수를 board에 저장한다
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                # 지뢰와 벽면의 경우에는 처리하지 않아야 한다
                if self.board[i][j] >= 0:
                    count = 0
                    # 적절한 처리를 통해 around를 사용하여 주변 8칸의 좌표를 구하고 순회한다
                    for k in range(8):
                        if self.board[i + int(around[2 * k + 1]) - 1][j + int(around[2 * k]) - 1] == -1:
                            count += 1
                    # 구한 주변 지뢰의 개수를 저장한다
                    self.board[i][j] = count

    # 우클릭 (메모) 시 작동하는 함수
    def click_right(self, x, y):
        # 벽면을 고려하여 값 보정
        x += 1
        y += 1
        # 이미 열린 칸은 변화하지 않는다
        if self.screen[y][x] == 1:
            return

        # 깃발이 없다면 세워주고 남은 지뢰의 개수를 하나 줄인다
        if self.screen[y][x] == 0:
            self.screen[y][x] = -1
            self.mine_left -= 1
            return

        # 깃발이 있다면 치워주고 남은 지뢰의 개수를 하나 늘린다
        self.mine_left += 1
        self.screen[y][x] = 0

    # 좌클릭 (열기) 시 작동하는 함수
    def click_left(self, x, y):
        # 벽면을 고려하여 값 보정
        x += 1
        y += 1
        # 닫힌 상태가 아닌 칸은 변화하지 않는다
        if self.screen[y][x]:
            return
        # 지뢰를 열었다면 게임을 오버시킨다
        if self.board[y][x] == -1:
            self.game_over()
            return
        # 지뢰가 아니라면 해당 칸을 search 해준다
        self.search(x, y)
        # 연 이후 승리했는지 확인
        self.check_win()

    # x,y 칸을 열고 만약 0이라면 주변도 열어주는 함수
    def search(self, x, y):
        # 현재 칸이 열려있거나 벽면이면 search하지 않는다
        if self.screen[y][x] == 1 or self.board[y][x] == -2:
            return
        # 현재 칸을 열어준다
        self.screen[y][x] = 1
        # 만약 현재 칸이 0이 아니라면 search 를 종료
        if self.board[y][x] != 0:
            return
        # 주변 8칸의 상대적인 좌표를 저장해주는 문자열
        around = "0001021012202122"
        # 현재 칸이 0이므로 적절한 처리를 통해 around를 사용하여 주변 8칸의 좌표를 구하고 해당 칸들도 search 해준다
        for k in range(8):
            self.search(x + int(around[2 * k]) - 1, y + int(around[2 * k + 1]) - 1)

    # 승리했는지 확인한다
    def check_win(self):
        # 열려 있는 칸의 갯수를 센다
        count = 0
        for i in self.screen:
            for j in i:
                if j == 1:
                    count += 1
        # 만약 지뢰를 제외한 모든 칸이 열려있다면 게임을 오버시키고 이겼음을 표시한다
        if count == self.left:
            self.is_game_over = True
            self.is_win = True

    # 게임 오버를 처리한다
    def game_over(self):
        # 게임 오버 상태로 표시한다
        self.is_game_over = True
        # 모든 칸을 돌며 지뢰를 표시해준다
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                if self.board[i][j] == -1:
                    self.screen[i][j] = 2

    # 보드를 프린트 해준다
    def print_board(self):
        res = "board:"
        for i in self.board:
            for j in i:
                if j != -2:
                    res += (" " if j >= 0 else "") + str(j) + " "
            res += '\n'
        print(res, end="")

    # 스크린을 프린트 해준다
    def print_screen(self):
        res = "screen:"
        for i in self.screen:
            for j in i:
                if j != -2:
                    res += (" " if j >= 0 else "") + str(j) + " "
            res += '\n'
        print(res, end="")

    # 실제로 보이는 화면을 프린트 해준다
    def print_real(self):
        # 남은 지뢰의 개수를 표시해준다
        print(f"left: {self.mine_left}", end="")
        for i in range(len(self.screen)):
            for j in range(len(self.screen[0])):
                # 벽면은 그리지 않는다
                if self.screen[i][j] == -2:
                    continue
                # 깃발은 p로 그려준다
                if self.screen[i][j] == -1:
                    print("p", end=" ")
                # 닫힌 칸은 o 로 그려준다
                if self.screen[i][j] == 0:
                    print("o", end=" ")
                # 지뢰는 * 로 그려준다
                if self.screen[i][j] == 2:
                    print("*", end=" ")
                # 열린 칸은 보드의 값을 불러와 그려준다
                if self.screen[i][j] == 1:
                    print(self.board[i][j], end=" ")
            print()
```

```python
# 크기 5*5짜리 지뢰 5개의 게임판 생성
g = Game(5, 5, 5)
# 디버깅을 위한 정답 보드를 프린트
g.print_board()

while True:
    # 게임 창을 출력한다
    g.print_real()
    # 게임이 종료되었는지 확인한다
    if g.is_game_over:
        # 이겼는지 오버된건지 확인하고 적절한 처리를 해준다
        if g.is_win:
            print("Win")
        else:
            print("Game over")
        break
    # 마우스 대신 숫자로 값을 입력 받는다
    # {마우스 클릭 0은 좌클릭 1은 우클릭} {x좌표(0부터 시작)} {y좌표(0부터 시작)}
    # ex) 0 2 2 는 (2,2)칸을 좌클릭 하는 것
    a, b, c = map(int, input().split())
    temp = [g.click_left, g.click_right][a]
    temp(b, c)
```
