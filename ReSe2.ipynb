```python
%reset -f
import matplotlib.pyplot as plt
# from matplotlib.colors import LogNorm
import numpy as np
import pandas as pd
from CTR import *
from CTR.PIticks import PIticks
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
# from CTR.Atoms import *
from CTR.Atoms import Re, Se
# from brokenaxes import brokenaxes

# import sys
# sys.path.append('../')
# from DFT import DFT 
```

```python
XRD = Xray(wavelength=1.5406)
XRD.Q(0,0,1)
# XRD2D = Xray2d()
# XRD2D.Q(0,0,1)
print(XRD.Energy)
```

```python
rese2 = Molecule(
    abc = (6.64, 6.77, 7.07),
    structure=(
        Re(0.064471, 0.315421, 0.491575), 
        Re(0.487198, 0.291296, 0.497211),
        Se(0.252667, 0.384044, 0.221090),
        Se(0.725641, 0.342656, 0.253293),
        Se(0.223326, 0.108908, 0.712059),
        Se(0.751740, 0.134689, 0.685320)
    )
)
# rese2.RJ

ReSe2 = Film(rese2, (1,1,1))
ReSe2_BULK = Film(rese2, (1,1,inf))
```

```python
# Sample(ReSe2, ReSe2_BULK, nref=(1,0,0))
(ReSe2/ReSe2_BULK(1,0,0)).I(XRD.q, XRD.Energy)
```

## Expand

```python
gen_ReSe2_film = lambda p : Film(rese2 * vdW(abc = (1,1,1+p/100)), (1,1,1))
ReSe2_EX_up    = gen_ReSe2_film(5)/ReSe2_BULK(1,0,0)
ReSe2_EX_down    = gen_ReSe2_film(-5)/ReSe2_BULK(1,0,0)


plt.figure(figsize=(7,4))
plt.semilogy(XRD.HKL(rese2)[:,2], XRD.I(ReSe2/ReSe2_BULK(1,0,0)), label='ยฑ0%', color = '#008060')
plt.semilogy(XRD.HKL(rese2)[:,2], XRD.I(ReSe2_EX_up), label='+5%', color = '#800020', linestyle='dashed')
plt.semilogy(XRD.HKL(rese2)[:,2], XRD.I(ReSe2_EX_down), label='$-$5%', color = '#006080', linestyle='dashed')

plt.legend()
plt.ylabel('Intensity (a.u.)', fontsize=20)  # Doubling the font size
plt.xticks(range(10), fontsize=16)  # Adjusting font size for x-ticks
plt.yticks(fontsize=16)
# plt.ylim([0, 0.25E10])
plt.xlabel("L (r.l.u.)", fontsize=20)  # Doubling the font size
plt.title("ReSe2 Bulk, (001) scan", fontsize=22)  # Doubling the font size
```
